```python
import pandas as pd         
import os                                                                                                              

sa_01_header = ['No','SubjectID', 'Housing complex', 'Gender', 'Birthday', 'Age', 
                'certification of care need', 'Number of residents', 'Height', 
                'Body weight', 'Stridecm)', 'Medical history', 'Disease name',
                'Regular hospital consultation', 'Antihypertensive', 
                'Antidepressant', 'Osteoporosis drug', 'Antidiabetic drug',
                'Other', 'Smoking', 'Drinking', 
                'Family composition of living together', 
                'Care need certification time', 'Hobbypreferencehabit', 
                'Activities', 'Community', 'Everyday', 
                'Everyday With or without blood pressure measurement', 
                'Whether or not to use a pedometer on a daily basis', 
                'Other2', 'right Measured value', 'right Evaluation', 
                'right Presence or absence of abnormality', 'left easured value',
                'left Evaluation', 'left Presence or absence of abnormality',
                'Locomotive Measured value', 
                'Locomotive Presence or absence of abnormality']

ir_04_header = ['No', 'SubjectID', 'Presence or absence of inspection record', 
                'Inspection date', 'Total cholesterol', 'LDL cholesterol', 
                'HDL cholesterol', 'Urea nitrogen', 'Creatinine',
                'Uric acid', 'Urine protein', 'Blood sugar', 'Diabetes diagnosis age',
                'HbA1c', 'AST', 'ALT', 'LDH', 'Retinopathy', 
                'Hypertension diagnosis age sodium', 'Na', 'K', 'ECG abnormalities',
                'Triglyceride', 'Diagnosis age of dyslipidemia', 
                'History of coronary artery disease', 'Chronic kidney disease diagnosis age',
                'Total protein', 'Albumin', 'Hematuria']

ev_05_header = ['No','SubjectID', 'Evaluation contents', 'Health status', 'Biological function', 
                'Exercise function', 'Nutrition', 'Cognitive function', 'Fall', 
                'High blood pressure', 'Arrhythmia', 'Osteoporosis', 
                'Diabetes mellitus', 'Exercise guidance', 'Nutrition guidance',
                'Life guidance']

li_02_header = ['No','SubjectID', 'Date confirmed', 'Appetite Questionnaire results',
                'Eating habit Questionnaire results', 'Preference Questionnaire results', 
                'Sleep Questionnaire results','Anxiety about health Questionnaire results', 
                'Anxiety about forgetfulness Questionnaire results', 
                'I can go up and down stairs without being transmitted to the railing or wall', 
                'I can get up from the chair', 'I can walk for more than 15 minutes', 
                'I can walk without a cane', 'I am actively going out', 
                'Walking function Score', 'Walking function Evaluation', 
                'Do the cleaning and washing yourself', 'Shop for daily necessities yourself',
                'I can prepare your own meals', 
                'I can write the documents to be submitted to the government office by myself', 
                '', ' Score', 
                ' Evaluation', 'I have a counselor in case of trouble',         
                "I have a counselor when I'm sick", 
                "I have someone who will take me to the hospital when I'm sick", 
                'There are people who take care of me when I fall asleep', 
                'Relying on someone', ' Score',  
                ' Evaluation', 
                'No weight loss of more than 2-3 kg in the last 6 months', 
                'No weight gain of more than 2-3 kg in the last 2 months', 
                'I can eat hard food',"When I drink tea or soup, I don't get sick", 
                "I don't mind thirst",
                ' Score', 
                ' Evaluation']
# transform
sa_01 = pd.read_csv('../../data/01_Subject_attributes.csv', header=[1])
sa_01.columns = sa_01_header
sa_01 = sa_01[sa_01['SubjectID'].notna()].reset_index(drop=True)

li_02 = pd.read_csv('../../data/02_Life_independence.csv', header=None)
li_02.columns = li_02_header
li_02 = li_02[li_02['SubjectID'].notna()].reset_index(drop=True)

ir_04 = pd.read_csv('../../data/04_Inspection_results.csv', header=None)
ir_04.columns = ir_04_header
ir_04 = ir_04[ir_04['SubjectID'].notna()].reset_index(drop=True)


ev_05 = pd.read_csv('../../data/05_evaluation.csv', header=None)
ev_05.columns = ev_05_header
ev_05 = ev_05[ev_05['SubjectID'].notna()].reset_index(drop=True)


# transform sensing data
f = []
for (dirpath, dirnames, filenames) in os.walk('../../data/71_sensing_data'):
  n = []
  for i in filenames:
    if i[-3:] != 'csv':
      continue
    n.append(os.path.join(dirpath , i))
  f.extend(n)

def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
s = set()
for i in f:
  c = i[45:]
  a = find_nth(c, '_', 1)
  b = find_nth(c, '_', 2)
  s.add(c[a+1:b])

for d in s:
  exec(d + " = pd.DataFrame()")

for file in f:
  c = file[45:]
  a = find_nth(c, '_', 1)
  b = find_nth(c[a+1:], '_', 1)
  if c[a+1:a+b+1] == 'activ':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    activ = activ.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'exist':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    exist = exist.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'sphy':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    sphy = sphy.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'temp':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    temp = temp.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'weight':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    weight = weight.append(tem, ignore_index=True)
    
activ['SubjectID'] = activ['SubjectID'] + '01'
exist['SubjectID'] = exist['SubjectID'] + '01'
sphy['SubjectID'] = sphy['SubjectID'] + '01'
temp['SubjectID'] = temp['SubjectID'] + '01'
weight['SubjectID'] = weight['SubjectID'] + '01'   

sphy = pd.DataFrame(sphy)
weight = pd.DataFrame(weight)
activ.to_csv('../../data/star_schema/activ.csv', index=False)
exist.to_csv('../../data/star_schema/exist.csv', index=False)
sphy.to_csv('../../data/star_schema/sphy.csv', index=False)
temp.to_csv('../../data/star_schema/temp.csv', index=False)
weight.to_csv('../../data/star_schema/weight.csv', index=False)
```
